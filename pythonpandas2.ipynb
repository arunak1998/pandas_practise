```python
df['BMI']=df['Weight']/((df['Height']/100)**2)
```

```python
df.head()
```

```python
bins = [0, 18.5, 24.9, 29.9, float('inf')]
labels = ['Underweight', 'Normal Weight', 'Overweight', 'Obesity']

# Use pd.cut to bin BMI values
df['BMI Category'] = pd.cut(df['BMI'], bins=bins, labels=labels, right=False)
```

```python
df.head()
```

```python
df['Stress Level '].unique()
```

```python
lable={
    'Bad':0, 'Awful':1, 'Good':2, 'fabulous':3
}
df['Stress Level ']=df['Stress Level '].map(lable)
```

```python
df['Stress Level ']=df['Stress Level '].astype(int)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.hist(df['Stress Level '], bins=6, color='blue', alpha=0.7)
plt.xlabel('StressLevel')
plt.ylabel('Frequency')
plt.title('Histogram of StressLevel Values')
plt.grid(True)
plt.show()
```
